## Importing essential Libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
```

## Data Preprocessing

```python
df=pd.read_csv('/content/drive/MyDrive/Churn_Modelling.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head().T
```

```python
new_df = df.drop(columns=['Surname', 'CustomerId', 'RowNumber'])
```

```python
new_df.info()
```

```python
new_df=pd.get_dummies(data=new_df,drop_first=True)
```

```python
new_df.head()
```

```python
new_df.shape
```

## Spliting the Data to training and testing sets and Feature Scaling

```python
x=new_df.drop('Exited',axis=1)
y=new_df['Exited']
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

## Builing the ANN model

```python
model=tf.keras.models.Sequential()
```

```python
#Adding input layer and first hidden layer
model.add(tf.keras.layers.Dense(units=6,activation='relu',input_dim=11))
```

```python
# Adding the 2nd hidden layer
model.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
## Output layer
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
# Compiling the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

## Training the Model

```python
model.fit(x_train,y_train.to_numpy(),batch_size=10,epochs=20)
```

## Evaluating the model

```python
test_loss,test_accuracy=model.evaluate(x_test,y_test.to_numpy())
```

```python
test_accuracy
```

```python
test_loss
```

```python
y_pred = np.argmax(model.predict(x_test), axis=1)
```
